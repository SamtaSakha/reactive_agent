```python
#reactive Agent: Thermostat
def reactive_agent(room_temp):
  if room_temp < 20:
    return "Heater On"
  elif room_temp > 25:
    return "Cooler On"
  else:
    return "Do Nothing"

temps =[18,22,27]
for temp in temps:
  action = reactive_agent(temp)
  print(f"Room Temperature: {temp}째C -> Action: {action}")
```
